```python
import torch

v1 = torch.arange(12)
print(v1.shape, v1.numel())
print(v1)
v2 = torch.tensor([[1],[2],[3]])
print(v2.shape, v2.numel())
v3 = torch.tensor([[1, 2, 3, 4]])
print(v3.shape, v3.numel())
v1 = v1.reshape(1,-1)
print(v1.shape, v1.numel())
print(v1)
v4 = torch.zeros(2, 3, 4)
print(v4.shape, v4.numel())
print(v4)
torch.manual_seed(7)
v5 = torch.randn(3,4)
print(v5.shape, v5.numel())
print(v5)
```
